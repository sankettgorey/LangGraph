```python
from langgraph.graph import START, END, StateGraph
from langgraph.graph.message import add_messages, BaseMessage
from langgraph.checkpoint.memory import InMemorySaver

from langchain_ollama import ChatOllama

from typing import TypedDict, List, Optional, Annotated

import time
```

```python
class AgentState(TypedDict):
    input: str
    step1: str
    step2: str
    step3: str
```

```python
print('hello')
```

```python
def step1(state: AgentState):

    print('Step 1 executed')

    return {
        'step1': 'Done', 
        # 'input': state['input']
    }

def step2(state: AgentState):

    time.sleep(30)
    print('Step 2 executed')

    return {
        'step2': 'Done', 
    }

def step3(state: AgentState):
    print('Step 3 executed')

    return {'step3': 'done'}
```

```python
checkpointer = InMemorySaver()
```

```python
graph = StateGraph(TypedDict)

graph.add_node('step1', step1)
graph.add_node('step2', step2)
graph.add_node('step3', step3)


graph.add_edge(START, 'step1')
graph.add_edge('step1', 'step2')
graph.add_edge('step2', 'step3')

workflow = graph.compile(checkpointer)
```

```python
config = {
    'configurable': {
        'thread_id': '1'
    }
}
```

```python
inital_state = {'input': 'start'}

workflow.invoke(inital_state, config)
```
